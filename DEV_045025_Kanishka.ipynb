```python
import numpy as np
```

```python
name= []
originalprice = []
offerprice = []
discount = []
offer = []
reviews=[]


for i in range(1,801):
    Url = 'https://www.nykaa.com/skin/c/8377?page_no={}&sort=popularity&eq=desktop'.format(i)

    all_pages = requests.get(Url)
    htmlCode = all_pages.text

    soup = BeautifulSoup(htmlCode, 'html.parser')


    for x in soup.find_all('div', attrs = {'class' : 'css-d5z3ro'}):

        #Productname
        product= x.find('div', attrs={'class':'css-xrzmfa'})
        if product is None:
            name.append(np.NaN)
        else:
            name.append(product.text)

        #reviews
        rating = x.find('span', attrs = {'class':'css-1qbvrhp'})
        if rating is None:
            reviews.append(np.NaN)
        else:
            reviews.append(rating.text)

         #originalprice
        orgprice =  x.find('span', attrs ={'class':'css-17x46n5'})
        if orgprice is None:
          originalprice.append(np.NaN)
        else:
            originalprice.append(orgprice.text)


        #offerprice
        off_price= x.find('span',{'class':'css-111z9ua'})
        if off_price is None:
            offerprice.append(np.NaN)
        else:
            offerprice.append(off_price.text)

        #discount
        disct = x.find('span',{'class':'css-cjd9an'})
        if disct is None:
            discount.append(np.NaN)
        else:
            discount.append(disct.text)


        #freegift
        gift = x.find('p',{'class':'css-1kzcg63'})
        if gift is None:
            offer.append(np.NaN)
        else:
            offer.append(gift.text)

print(len(name))
print(len(originalprice))
print(len(offerprice))
print(len(discount))
print(len(offer))
print(len(reviews))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Product Name': name, 'Reviews' : reviews, 'Original Price' : originalprice, 'Offer Price' : offerprice, 'Discount': discount, 'Free Gifts': offer})
print(df.shape)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'/content/Nykaa_BestSkinProducts.csv')
df.head()
```

```python
#Dataset information in rows and columns
df.shape
```

```python
#Complete information of the dataset
df.info()
```

```python
df.columns
```

```python
#Dropping unwanted columns
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
#Cleaning Columns
df['Reviews'] = df.Reviews.str.replace('\(|\)', '', regex = True)
df['Original Price'] = df['Original Price'].str.replace('MRP:₹', '', regex = True)
df['Original Price'] = df['Original Price'].str.replace('MRP:', '', regex = True)
df['Offer Price'] = df['Offer Price'].str.replace('₹', '', regex=True)
```

```python
df.head()
```

```python
#checking null values
df.isnull().sum()
```

```python
#Let's fill last two columns
df.Discount = df.Discount.fillna('0% Off')
df['Free Gifts'] = df['Free Gifts'].fillna('No Gift')
df.info()
```

```python
#changing dtypes of several columns from object to numeric
df['Reviews'] = pd.to_numeric(df['Reviews'])
df['Original Price'] = pd.to_numeric(df['Original Price'])
df['Offer Price'] = pd.to_numeric(df['Offer Price'])
df.info()
```

```python
#Now we can fill the missing values in Original Price with Offer Price
df['Original Price'] = df['Original Price'].fillna(df['Offer Price'])


#Filling missing values in Reviews column with 0
df.Reviews = df.Reviews.fillna(0)
df.info()
```

```python
#Checking null values after filling the data
df.isnull().sum()
```

```python
#Checking duplicate values
df.duplicated().sum()
```

```python
#Checking duplicates in a tabular format for a better understanding
duplicate_rows = df.duplicated()
duplicate_data = df[duplicate_rows]
duplicate_data.sample(10)
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#Let's extract the Brand name from the Product name column
df['Product Name'].sample(10)
```

```python
col_split = df['Product Name'].str.split('/', n=0, expand=True)
col_split

#Importing regular expression library
import re

df['Brand'] = df['Product Name'].apply(lambda x: ' '.join(re.findall(r'\b\w+\b', x)[:3]) if re.findall(r'\b\w+\b', x) else None)

# Print the extracted brand names
print(df['Brand'])
```

```python
df.head()
```

```python
df['Discount (% OFF)'] = df['Discount'].str.replace('% Off', '', regex=True)
df['Discount (% OFF)'] = pd.to_numeric(df['Discount (% OFF)'])
df.head()
```

```python
# Calculate the price difference
df['Price Difference'] = df['Original Price'] - df['Offer Price']

# Sort the DataFrame by Price Difference column
sorted_df = df.sort_values('Price Difference', ascending=False)

# Plot the price difference analysis using seaborn
plt.figure(figsize=(10, 6))
sns.countplot(x='Price Difference', data=sorted_df.head(10), palette='viridis')
plt.title('Highest Price Difference')
plt.xlabel('Price Difference')
plt.xticks(rotation=45)
plt.show()

```

```python
# Plot the price difference analysis using seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for illustration purposes
data = {'Product Name': ['Product A', 'Product B', 'Product C'],
        'Price Difference': [10, 20, 15]}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort the DataFrame by 'Price Difference' in descending order
sorted_df = df.sort_values(by='Price Difference', ascending=False)

# Plot the price difference analysis using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Price Difference', y='Product Name', data=sorted_df.head(10), palette='viridis')
plt.title('Top 10 Products with the Highest Price Difference')
plt.ylabel('Product Name')
plt.show()
```
